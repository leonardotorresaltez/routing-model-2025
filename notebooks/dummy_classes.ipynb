```python
from typing import List, Tuple, Dict
from dataclasses import dataclass, field
```

```python
@dataclass
class Customer:
    id: int
    lat: float
    lon: float
    road_access_type: str = 'A'
    client_type: str = 'standard'
    # optional
    route_times: Dict[str, float] = field(default_factory=dict) # distance from customer1 to customer2

    def location(self) -> Tuple[float, float]:
        return (self.lat, self.lon)

@dataclass
class Truck:
    id: int
    depot_id: int

@dataclass
class Depot:
    id: int
    lat: float
    lon: float
    truck_fleet: List[int] = field(default_factory=list)
    # optional
    route_times: Dict[str, float] = field(default_factory=dict) # distance from depot to customer2
    
    def location(self) -> Tuple[float, float]:
        return (self.x, self.y)

@dataclass
class TruckState:
    truck: Truck
    current_location: Tuple[float, float]
    visited_customers: List[int] = field(default_factory=list)
    pending_customers: List[int] = field(default_factory=list)
    distance_traveled: float
    
    
@dataclass
class Product:
    id: int
    weight: float
    height: float
    length: float
    width: float
    volume: float = field(init=False)

    def __post_init__(self):
        self.volume = self.weight * self.length * self.width
        
        
@dataclass
class Pallet:
    id: int
    type: int
    max_weight: float
    base_weight: float
    height: float
    length: float
    width: float
    volume: float = field(init=False)

    def __post_init__(self):
        self.volume = self.weight * self.length * self.width
```
